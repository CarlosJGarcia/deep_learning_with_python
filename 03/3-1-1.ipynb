```python
import tensorflow as tf
from tensorflow import keras
from keras import models, layers, optimizers

from keras.datasets import mnist
```

```python
print("1. Carga de datos")
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

print()
print(f"Clase:{type(train_images)}")
print("Orignal:")
print(f"- Dimensión: {train_images.ndim}D")
print(f"- Shape: {train_images.shape} - (muestras, x, y)")
print(f"- Clase de los items: {train_images.dtype}")

# De array 3D (60000, 28, 28) a 2D (60000, 28*28). 28*28=786 es el número de neuronas en la capa de entrada.
# Typecasting de entero 8 bits a punto flotante 32 bits. Es un requerimiento del algoritmo de backpropagation
# Normalización de rango 0-255 a 0-1. Ayuda a que el modelo 'converja' antes
train_images = train_images.reshape(60000, 28*28)
train_images = train_images.astype('float32')/255

print()
print("Tras reshape, typecasting y normalización:")
print(f"- Dimensión: {train_images.ndim}D")
print(f"- Shape: {train_images.shape} - (muestras, caracteristicas)")
print(f"- Clase de los items: {train_images.dtype}")
```

```python
model = models.Sequential()

model.add(layers.Input(shape=(784,)))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

model. compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print("2. Modelo de dos capas definido correctamente")
```

Modelo secuencial de dos capas densamente conectadas
Entrada: tensor 2D (muestras, caracteristicas) = (60000, 784)
Salida: tensor 1D (muestras, caracteristicas) = (60000, 10)

```python
print("3. Training")
model.fit(train_images, train_labels, batch_size=128, epochs=10)
```

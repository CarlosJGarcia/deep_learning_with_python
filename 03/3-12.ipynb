```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

from keras.datasets import reuters
```

```python
(train_data, train_labels), (test_data, test_labels) = reuters.load_data(num_words=10000)
```

```python
print("Datos entrenamiento:  ", len(train_data))
print("Datos de prueba:      ", len(test_data))

unique_values = np.unique(train_labels)
print("Número de categorías: ", len(unique_values))
```

```python
# Traduzco a palabras

word_index = reuters.get_word_index()
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
decoded_news = ' '.join(reverse_word_index.get(i - 3, '?') for i in train_data[0])
print(decoded_news)
print ("Etiqueta: ", train_labels[0])
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import models, layers, optimizers

from keras.datasets import imdb
```

```python
print("1. Carga de datos")
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)

print()
print("train_data:")
print(f"- Clase:{type(train_data)}")
print(f"- Dimensión: {train_data.ndim}D")
print(f"- Shape: {train_data.shape} - (muestras)")
item = train_data[0]
print(f"- Clase de los items:{type(item)}")

for n in range(0,10):
    item = train_data[n]
    print(f"   - Item(lista) {n} tiene {len(item)} 'palabras'")

print()
print("train_labels:")
print(f"- Clase:{type(train_labels)}")
print(f"- Dimensión: {train_labels.ndim}D")
print(f"- Shape: {train_labels.shape} - (muestras, x, y)")
print(f"- Clase de los items: {train_labels.dtype}")

for n in range(0,10):
    item = train_labels[n]
    print(f"   - Label {n}: {train_labels[n]}")

```

```python
# 1. Load the dictionary (Word -> Index)
word_index = imdb.get_word_index()

# 2. Reverse it (Index -> Word) and account for the offset
# We add 3 because the actual words start at index 3 in the training data
reverse_word_index = {value + 3: key for (key, value) in word_index.items()}

# 3. Manually define the reserved tokens
reverse_word_index[0] = "<PAD>"
reverse_word_index[1] = "<START>"
reverse_word_index[2] = "<OOV>"
reverse_word_index[3] = "<UNUSED>"

# 4. Decode the first review
n = 0
review = train_data[n]
decoded_review = ' '.join([reverse_word_index.get(i, '?') for i in review])
print(f"Review número {n}")
print("Decoded review:")
print(decoded_review)
print(f"Número de palabras: {len(review)}")
print(f"Crítica: {train_labels[n]}")
if train_labels[n] == 1:
    print("Crítica positiva")
else:
    print("Crítica negativa")
```

El método imdb.get_word_index() devuelve el diccionario palabras-token del dataset Keras IMDB
Es un diccionario python del tipo:

word_index = {'perro': 500, 'gato': 333, 'hola': 25}

El valor numérico de cada token corresponde al número de veces que cada palabra aparece en el dataset

```python
import tensorflow as tf
from tensorflow import keras
from keras.datasets import imdb

print("Cargando keras.IMDB")
# (train_data, train_labels), (test_data, test_labels) = imdb.load_data()
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)
print("Hecho")
```

```python
print("Número de GPUs: ", len(tf.config.list_physical_devices('GPU')))
```

```python
import sys 

def get_deep_size(obj):
    """Calculates the real memory footprint of an object and its contents."""
    size = sys.getsizeof(obj)
    
    # If the object is a collection, add the size of its items
    if isinstance(obj, (list, tuple, set, frozenset)):
        for item in obj:
            size += get_deep_size(item)
    elif isinstance(obj, dict):
        # For dicts, we need to count both keys and values
        for k, v in obj.items():
            size += get_deep_size(k)
            size += get_deep_size(v)
            
    return size

train = get_deep_size((train_data, train_labels))
test = get_deep_size((test_data, test_labels))
kb = (train + test) // 1024
print(f"Size in memory {kb} KB")
```
